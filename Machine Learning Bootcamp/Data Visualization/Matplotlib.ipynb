```python
import numpy as np
import matplotlib.pyplot as plt

# Shows plot inline in Jypyter
%matplotlib inline
```

```python
x = np.linspace(0, 5, 11)
y = x ** 2
```

## Functional method to create plot

```python
plt.plot(x, y)
# plt.show()   If not using JN
plt.xlabel('x-label')
plt.ylabel('y-label')
plt.title('Title')
```

## Mutliple plots on same canvas

```python
plt.subplot(1, 2, 1)
plt.plot(x, y, 'r')
plt.subplot(1, 2, 2)
plt.plot(y, x, 'b')
```

## Object oriented method
Prefered method

```python
fig = plt.figure()
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes.plot(x, y)
axes.set_xlabel('X-label')
axes.set_ylabel('Y-label')
axes.set_title('Title')
```

```python
# 1. Create empty canvas
fig = plt.figure()

# 2. Add axes
axes1 = fig.add_axes([0.1, 0.1, 0.7, 0.8])
axes2 = fig.add_axes([1.0, 0.1, 0.4, 0.5])

# 3. Plot data
axes1.plot(x, y, 'r')
axes2.plot(y, x, 'b')
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)
for c in axes:
    c.plot(x, y)
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)
axes[0].plot(x, y)
axes[1].plot(y, x)
```

## Figure Size and DPI

```python
fig = plt.figure(figsize=(4, 5), dpi=100)
ax = fig.add_axes([0, 0, 1,1])
ax.plot(x, y)
```

```python
fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(8, 3))
axes[0].plot(x, y)
axes[1].plot(y, x)
plt.tight_layout()
```

```python
fig.savefig('MY_GRAF.png', dpi=200)
```

## Two lines in same graf

```python
fig = plt.figure()
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes.plot(x, y, 'r', label='First')
axes.plot(y, x**2, 'g', label='Second')
axes.legend(loc=0)
```

## Plot appearance

```python
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, y, color='orange', linewidth=3, alpha=0.7, 
        linestyle='--', marker='o', 
        markersize=20, markerfacecolor='purple', markeredgewidth=3)
```

```python
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, y, color='orange', linewidth=3, alpha=0.7, 
        linestyle='--')
ax.set_xlim([0, 1])
ax.set_ylim([0, 2])
```
