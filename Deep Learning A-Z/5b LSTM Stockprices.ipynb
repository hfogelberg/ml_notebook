```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from keras.models import Sequential
from keras.layers import Dense, LSTM
```

## Get the data

```python
ds = pd.read_csv('data/Google_Stock_Price_Train.csv')
ds.head()
# Only need the open column
ds = ds.iloc[:, 1:2].values
```

```python
# Feature scaling using normalization
sc = MinMaxScaler()
ds = sc.fit_transform(ds)
```

```python
# Exclude the last line, since we try to predict the value for the next day
# There are 1258 lines of data in the file
X_train = ds[0:1257]

# Y-train is X-train shifted by one
y_train = ds[1:1258]
```

## Preprocessing
### Reshape
Shape the format of X_train to a 3D array:
- observations: 1257
- timestep: 1
- feature: 1

This is used as input to Keras.

```python
X_train = np.reshape(X_train, (1257, 1, 1))
```

## Create the RNN

```python
# Initiate
regressor = Sequential()

# Add LSTM layer
regressor.add(LSTM(units=4, activation='sigmoid', input_shape=(None, 1)))

# Output layer
regressor.add(Dense(units=1))

# Compile
regressor.compile(optimizer='adam', loss='mean_squared_error')

# Fit model to training set
regressor.fit(X_train, y_train, batch_size=32, epochs=200)
```
